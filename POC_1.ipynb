# **KNN classifier**

```python
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1,y1,test_size=0.30,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=9, p=2)
classifier.fit(x1_train, y1_train)
```

```python
y1_pred = classifier.predict(x1_test)
```

```python
from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y1_test, y1_pred)
```

```python
cm1
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y1_test, y1_pred)
precision = precision_score(y1_test, y1_pred, average='macro')
recall = recall_score(y1_test, y1_pred, average='macro')
f1 = f1_score(y1_test, y1_pred, average='macro')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("f1 Score:", f1)
```
